# Exploratory data analysis
 

```python
df.columns
```

```python
ax =sns.countplot(x='Gender',data =df)
```

```python
ax =sns.countplot(x='Gender',data =df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen
```

```python
sns.barplot(x= 'Gender', y='Amount',data= sales_gen)
plt.show()
```

#from above graphs we can see that most of the buyers are females and even the purchasing poer of females are greater than men

# Age


```python
df.columns
```

```python
ax=sns.countplot(data=df, x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount vs age group
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
plt.show()
```

# states


```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x='State',y='Orders',data=sales_state)
plt.show()
```

```python
#total number of amounts/sales from top 10 states
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(x='State',y='Amount',data=sales_state)
plt.show()
```

#from above graphs we can see that most of the orders are from uttar pradesh,maharashtra and karnataka respectively

# Maritial status

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(6,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,3)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender')
```

# from the above graphs we can see that most of the buyers are married(women) and they have high purchasing power
